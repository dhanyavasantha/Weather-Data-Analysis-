```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
data = pd.read_csv("weatherHistory.csv")
```

```python
data
```

```python
data.describe()
```

```python
data = data.dropna()
```

```python
data['Formatted Date'] = pd.to_datetime(data['Formatted Date'])
```

```python
data = data.drop_duplicates()
```

```python
data = data.drop(columns=['Loud Cover'])
```

```python
data
```

```python
data.describe()
```

```python
data.to_csv("weather_data.csv", index=False)
```

```python
# Plotting temperature over time
plt.figure(figsize=(10, 6))
plt.plot(data['Formatted Date'], data['Temperature (C)'], color='blue')
plt.title('Temperature Variation Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plotting humidity over time
plt.figure(figsize=(10, 6))
plt.plot(data['Formatted Date'], data['Humidity'], color='green')
plt.title('Humidity Variation Over Time')
plt.xlabel('Date')
plt.ylabel('Humidity')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plotting pressure over time
plt.figure(figsize=(10, 6))
plt.plot(data['Formatted Date'], data['Pressure (millibars)'], color='red')
plt.title('Pressure Variation Over Time')
plt.xlabel('Date')
plt.ylabel('Pressure (millibars)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Get unique values of the 'Daily Summary' column
unique_summaries = data['Summary'].unique()

# Print the unique values
print(unique_summaries)
print(len(unique_summaries))
```

```python
summary_temp = data.groupby('Summary')['Temperature (C)'].mean().reset_index()
# Plotting summary vs. mean temperature
plt.figure(figsize=(12, 6))
plt.bar(summary_temp['Summary'], summary_temp['Temperature (C)'], color='red')
plt.title('Mean Temperature for Each Summary')
plt.xlabel('Summary')
plt.ylabel('Mean Temperature (C)')
plt.xticks(rotation=90, fontsize=8)
plt.tight_layout()
plt.show()
```

```python
summary_hum = data.groupby('Summary')['Humidity'].mean().reset_index()
# Plotting summary vs. mean temperature
plt.figure(figsize=(12, 6))
plt.bar(summary_hum['Summary'], summary_hum['Humidity'], color='red')
plt.title('Mean Humidity for Each Summary')
plt.xlabel('Summary')
plt.ylabel('Mean Temperature (C)')
plt.xticks(rotation=90, fontsize=8)
plt.tight_layout()
plt.show()
```
